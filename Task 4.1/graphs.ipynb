```python
import pandas as pd
import numpy as num
import matplotlib.pyplot as plt
from google.colab import files
import seaborn as sb

```

```python
weather = pd.read_csv("/content/drive/MyDrive/Weather.csv")


weather = weather.sort_values(by="Formatted Date",ascending=True)
weather = weather[weather["Humidity"] >=0]
weather = weather[weather["Temperature (C)"] > -40]
weather = weather.drop_duplicates()

plt.plot(weather["Formatted Date"],weather["Apparent Temperature (C)"],label='Temperature (C)')
plt.xticks(weather["Formatted Date"][::500],  rotation='vertical')



plt.xlabel("Time")
plt.ylabel("Temperature (C)")
plt.legend()
plt.show()

weather.head(310)
```

```python
plt.hist(weather["Apparent Temperature (C)"],edgecolor="black")
plt.xlabel("Temperature")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.scatter(weather["Temperature (C)"],weather["Humidity"],s=1)

plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.show()
```

```python
weatherheat = weather.drop(columns=['Daily Summary','Precip Type','Summary','Formatted Date'])
sb.heatmap(weatherheat.corr(),annot=True)
plt.show()
```

**Humidity seems to be inversely proportional with temperature to a great degree**
